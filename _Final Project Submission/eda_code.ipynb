```python
# 5. OUTLIER DETECTION using IQR method
print("=" * 70)
print("5. OUTLIER DETECTION (IQR Method)")
print("=" * 70)

# Focus on key clinical variables
key_numeric = [c for c in numeric_cols if c not in ['subject_id', 'hadm_id']]

outlier_summary = []
for col in key_numeric:
    data = df_merged[col].dropna()
    if len(data) == 0:
        continue
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    n_outliers = ((data < lower) | (data > upper)).sum()
    pct_outliers = round(n_outliers / len(data) * 100, 2)
    outlier_summary.append({
        'Column': col,
        'N': len(data),
        'Q1': round(Q1, 3),
        'Q3': round(Q3, 3),
        'IQR': round(IQR, 3),
        'Lower Fence': round(lower, 3),
        'Upper Fence': round(upper, 3),
        'N Outliers': n_outliers,
        'Outlier %': pct_outliers
    })

outlier_df = pd.DataFrame(outlier_summary).sort_values('Outlier %', ascending=False)
print(outlier_df[outlier_df['N Outliers'] > 0].to_string(index=False))
print(f"\nColumns with no outliers: {(outlier_df['N Outliers'] == 0).sum()}")
```

```python
# 6. DISTRIBUTION ANALYSIS - Skewness and Kurtosis
print("=" * 70)
print("6. DISTRIBUTION ANALYSIS - Skewness and Kurtosis")
print("=" * 70)

from scipy import stats

dist_summary = []
for col in key_numeric:
    data = df_merged[col].dropna()
    if len(data) < 8:
        continue
    sk = stats.skew(data)
    kt = stats.kurtosis(data)
    # Shapiro-Wilk test (use sample if n > 5000)
    if len(data) <= 5000:
        sw_stat, sw_p = stats.shapiro(data)
    else:
        sw_stat, sw_p = np.nan, np.nan
    
    dist_summary.append({
        'Column': col,
        'N': len(data),
        'Skewness': round(sk, 3),
        'Kurtosis': round(kt, 3),
        'Shapiro W': round(sw_stat, 4) if not np.isnan(sw_stat) else np.nan,
        'Shapiro p': sw_p,
        'Normal (p>0.05)': 'Yes' if sw_p > 0.05 else 'No'
    })

dist_df = pd.DataFrame(dist_summary)
print(dist_df.to_string(index=False))

print(f"\n\nNormally distributed (Shapiro p > 0.05): {(dist_df['Normal (p>0.05)'] == 'Yes').sum()} / {len(dist_df)}")
print(f"Non-normal: {(dist_df['Normal (p>0.05)'] == 'No').sum()} / {len(dist_df)}")
```

```python
# Key continuous variables to test across groups
tukey_targets = [
    'age', 'baseline_creatinine', 'map_twa_24h', 'map_min_24h', 'map_variance',
    'hr_admit', 'pressor_hours', 'max_pressor_dose', 'urine_output_24h',
    'fluid_balance_24h', 'charlson_score', 'elixhauser_score',
    'lactate_mean', 'lactate_max', 'bun_mean', 'sodium_mean', 'potassium_mean',
    'glucose_mean', 'wbc_mean', 'platelets_mean', 'hr_mean',
    'peak_troponin_24h', 'peak_troponin_7day',
    'min_hgb_24h', 'icu_los_days', 'hospital_los_days'
]

# ============================================================
# TUKEY HSD by PROCEDURE GROUP (ENDO vs OPEN)
# ============================================================
print("=" * 70)
print("TUKEY HSD: procedure_group (ENDO vs OPEN)")
print("=" * 70)

for var in tukey_targets:
    data = df_merged[['procedure_group', var]].dropna()
    if len(data) < 10:
        continue
    
    result = pairwise_tukeyhsd(data[var], data['procedure_group'], alpha=0.05)
    # Only print if significant
    sig = result.reject[0]
    meandiff = result.meandiffs[0]
    pval = result.pvalues[0]
    ci_low = result.confint[0][0]
    ci_high = result.confint[0][1]
    
    marker = "***" if pval < 0.001 else "**" if pval < 0.01 else "*" if pval < 0.05 else ""
    if sig:
        print(f"\n{var} {marker}")
        print(f"  Mean diff (OPEN - ENDO): {meandiff:.4f}, p={pval:.6f}")
        print(f"  95% CI: [{ci_low:.4f}, {ci_high:.4f}]")
        # Group means
        grp_means = data.groupby('procedure_group')[var].agg(['mean', 'std', 'count'])
        print(f"  ENDO: mean={grp_means.loc['ENDO','mean']:.3f} (SD={grp_means.loc['ENDO','std']:.3f}, n={int(grp_means.loc['ENDO','count'])})")
        print(f"  OPEN: mean={grp_means.loc['OPEN','mean']:.3f} (SD={grp_means.loc['OPEN','std']:.3f}, n={int(grp_means.loc['OPEN','count'])})")

print("\n\n(Only significant results shown. Non-significant variables omitted.)")
```

```python
# ============================================================
# TUKEY HSD by BASELINE CKD CLASS (3 groups)
# ============================================================
print("=" * 70)
print("TUKEY HSD: baseline_class_id (CKD Baseline Stage 1, 2, 3)")
print("=" * 70)

for var in tukey_targets:
    data = df_merged[['baseline_class_id', var]].dropna()
    if len(data) < 10:
        continue
    
    # Check minimum group sizes
    grp_sizes = data.groupby('baseline_class_id')[var].count()
    if grp_sizes.min() < 5:
        continue
    
    result = pairwise_tukeyhsd(data[var], data['baseline_class_id'], alpha=0.05)
    
    # Check if any comparison is significant
    any_sig = any(result.reject)
    if any_sig:
        print(f"\n{'='*50}")
        print(f"{var}")
        print(f"{'='*50}")
        grp_stats = data.groupby('baseline_class_id')[var].agg(['mean', 'std', 'count'])
        for idx in grp_stats.index:
            print(f"  Class {idx}: mean={grp_stats.loc[idx,'mean']:.3f} (SD={grp_stats.loc[idx,'std']:.3f}, n={int(grp_stats.loc[idx,'count'])})")
        
        print(f"\n  Pairwise comparisons:")
        summary = result.summary()
        for i in range(len(result.reject)):
            g1 = result._results_table[i+1][0]
            g2 = result._results_table[i+1][1]
            md = result.meandiffs[i]
            p = result.pvalues[i]
            rej = result.reject[i]
            marker = "***" if p < 0.001 else "**" if p < 0.01 else "*" if p < 0.05 else "ns"
            print(f"    {g1} vs {g2}: diff={md:.4f}, p={p:.6f} {marker}")
```

```python
# ============================================================
# TUKEY HSD by MAP CATEGORY (3 groups, excluding NaN)
# ============================================================
print("=" * 70)
print("TUKEY HSD: map_category (Shock <65, Standard 65-74.9, Aggressive >=75)")
print("=" * 70)

for var in tukey_targets:
    data = df_merged[['map_category', var]].dropna()
    if len(data) < 10:
        continue
    
    grp_sizes = data.groupby('map_category')[var].count()
    if grp_sizes.min() < 5:
        continue
    
    result = pairwise_tukeyhsd(data[var], data['map_category'], alpha=0.05)
    
    any_sig = any(result.reject)
    if any_sig:
        print(f"\n{'='*50}")
        print(f"{var}")
        print(f"{'='*50}")
        grp_stats = data.groupby('map_category')[var].agg(['mean', 'std', 'count'])
        for idx in grp_stats.index:
            print(f"  {idx}: mean={grp_stats.loc[idx,'mean']:.3f} (SD={grp_stats.loc[idx,'std']:.3f}, n={int(grp_stats.loc[idx,'count'])})")
        
        print(f"\n  Pairwise comparisons:")
        for i in range(len(result.reject)):
            g1 = result._results_table[i+1][0]
            g2 = result._results_table[i+1][1]
            md = result.meandiffs[i]
            p = result.pvalues[i]
            marker = "***" if p < 0.001 else "**" if p < 0.01 else "*" if p < 0.05 else "ns"
            print(f"    {g1} vs {g2}: diff={md:.4f}, p={p:.6f} {marker}")
```

```python
# Fix: g1/g2 might not be simple numbers, let's just print them as-is
print("=" * 70)
print("TUKEY HSD: max_aki_stage_7day (0, 1, 2, 3)")
print("=" * 70)

for var in tukey_targets:
    data = df_merged[['max_aki_stage_7day', var]].dropna()
    if len(data) < 10:
        continue
    
    grp_sizes = data.groupby('max_aki_stage_7day')[var].count()
    if grp_sizes.min() < 5:
        continue
    
    result = pairwise_tukeyhsd(data[var], data['max_aki_stage_7day'], alpha=0.05)
    
    any_sig = any(result.reject)
    if any_sig:
        print(f"\n{'='*50}")
        print(f"{var}")
        print(f"{'='*50}")
        grp_stats = data.groupby('max_aki_stage_7day')[var].agg(['mean', 'std', 'count'])
        for idx in grp_stats.index:
            print(f"  Stage {int(idx)}: mean={grp_stats.loc[idx,'mean']:.3f} (SD={grp_stats.loc[idx,'std']:.3f}, n={int(grp_stats.loc[idx,'count'])})")
        
        print(f"\n  Significant pairwise comparisons:")
        for i in range(len(result.reject)):
            md = result.meandiffs[i]
            p = result.pvalues[i]
            if p < 0.05:
                ci = result.confint[i]
                marker = "***" if p < 0.001 else "**" if p < 0.01 else "*" if p < 0.05 else "ns"
                # Extract group labels from summary table
                row = result.summary().data[i+1]
                print(f"    {row[0]} vs {row[1]}: diff={md:.4f}, p={p:.6f}, 95% CI=[{ci[0]:.4f}, {ci[1]:.4f}] {marker}")
```

```python
# 13. EFFECT SIZE ANALYSIS (Cohen's d) for key comparisons
print("=" * 70)
print("13. EFFECT SIZE ANALYSIS (Cohen's d) - Procedure Group")
print("=" * 70)

def cohens_d(group1, group2):
    n1, n2 = len(group1), len(group2)
    var1, var2 = group1.var(), group2.var()
    pooled_std = np.sqrt(((n1-1)*var1 + (n2-1)*var2) / (n1+n2-2))
    if pooled_std == 0:
        return 0
    return (group1.mean() - group2.mean()) / pooled_std

print(f"\n{'Variable':<30} {'Cohen d':>10} {'Interpretation':>15}")
print("-" * 60)
for var in tukey_targets:
    data = df_merged[['procedure_group', var]].dropna()
    endo = data[data['procedure_group'] == 'ENDO'][var]
    opn = data[data['procedure_group'] == 'OPEN'][var]
    if len(endo) < 5 or len(opn) < 5:
        continue
    d = cohens_d(endo, opn)
    interp = "negligible" if abs(d) < 0.2 else "small" if abs(d) < 0.5 else "medium" if abs(d) < 0.8 else "large"
    if abs(d) >= 0.2:  # Only show non-negligible effects
        print(f"{var:<30} {d:>10.3f} {interp:>15}")
```
